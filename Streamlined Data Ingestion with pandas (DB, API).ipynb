```python
query_10 = """
    SELECT borough, COUNT(*) AS plumbing_complaints
    FROM hpd311calls
    WHERE complaint_type = 'PLUMBING'
    GROUP BY borough
"""
plumbing_by_borough = pd.read_sql(query_10, sqlite_engine)
plumbing_by_borough
```

```python
# JOIN ... ON ... must be same data type
query_11 = """
    SELECT unique_key, weather.date, incident_address, borough
    FROM hpd311calls
    JOIN weather
    ON hpd311calls.created_date = weather.date
    WHERE hpd311calls.complaint_type = 'HEAT/HOT WATER'
"""
calls_and_weathers_hot_water = pd.read_sql(query_11, sqlite_engine)
calls_and_weathers_hot_water
```

```python
query_12 = """
    SELECT hpd311calls.borough, COUNT(*), boro_census.total_population, boro_census.housing_units
    FROM hpd311calls
    JOIN boro_census
    ON hpd311calls.borough = boro_census.borough
    GROUP BY hpd311calls.borough
"""
population_borough = pd.read_sql(query_12, sqlite_engine)
population_borough
```

```python
# Using Yelp API
business_search_url = 'https://api.yelp.com/v3/businesses/search'
params = {'term': 'bookstore', 'location': 'San Francisco'}
headers = {'Authorization': 'Bearer {}'.format(yelp_api_key)}

response = requests.get(business_search_url, params=params, headers=headers)
yelp_data = response.json()

bookstores = pd.DataFrame(yelp_data['businesses'])
print(bookstores.head(2))
```

```python
bookstores_meta = pd.json_normalize(yelp_data["businesses"], 
                                    sep="_", record_path="categories", 
                                    meta=["name", "alias", "rating", 
                                          ["coordinates", "latitude"], 
                                          ["coordinates", "longitude"]], 
                                    meta_prefix="biz_")
bookstores_meta
```
